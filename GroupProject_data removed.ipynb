```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as patches
from astropy.table import Table
from scipy.optimize import curve_fit
star_table = Table.read('Desktop/ASTR19/Planet_Lightcurve.fits')
print(star_table["time [days]"][0:416])
```

```python
x1 = star_table["time [days]"]
y1 = star_table["flux"]
plt.plot(x1, y1)

x_poly = [-0.04, 0.04, 0.04, -0.04, -0.04]
y_poly = [1.01, 1.01, 0.965, 0.965, 1.01]

plt.plot(x_poly, y_poly, "-o", c="r")

coords = np.column_stack((x_poly, y_poly))

poly = patches.Polygon(coords)

points_coords = np.column_stack((x1,y1))

wd_mask = poly.contains_points(points_coords)

outside_mask = ~wd_mask
```

```python
plt.plot(x1[outside_mask], y1[outside_mask], '.')
```
